```python
import pandas as pd
import shutil
import os
import sys
import cv2
import matplotlib.pyplot as plt
import random
```

```python
from __future__ import print_function
labels = pd.read_csv(r'/Volumes/Extreme SSD/DR/trainLabels.csv')
train_dir =r'/Volumes/Extreme SSD/DR/CLAHE_1'
DR = r"/Volumes/Extreme SSD/DR/CLAHE_new"
if not os.path.exists(DR):
    os.mkdir(DR)

for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(DR + str(class_name)):
        os.mkdir(DR + str(class_name))
    src_path = train_dir + '/'+ filename + '.jpeg'
    dst_path = DR + str(class_name) + '/' + filename + '.jpeg'
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```

```python
import os, os.path

class_fol = '/Volumes/Extreme SSD/DR/CLAHE_new/'
sub_fol = [ f.path for f in os.scandir(class_fol) if f.is_dir()]
print(sub_fol)

for subfolder in sub_fol:
    length = len([name for name in os.listdir(subfolder) if os.path.isfile(os.path.join(subfolder,name))])
    print(length)
```

```python
path = r'/Volumes/Extreme SSD/DR/CLAHE_new/0'# You can provide the path here
tot_no = len(os.listdir(path))
print("Total number of images in the folder:",tot_no)
n = tot_no - 10000 # Number of random images to be removed
print("Total number of images removed:",n)
img_names = os.listdir(path)  # Get image names in folder
img_names = random.sample(img_names, n)  # Pick 2500 random images
for image in img_names:  # Go over each image name to be deleted
    f = os.path.join(path, image)  # Create valid path to image
    #print(f)
    #break
    os.remove(f)  # Remove the image
```
